```python
import numpy as np
pv=df.pivot_table(["total"],index="sku",columns="quantity",margins=True,aggfunc=np.sum,fill_value=0) #数据透视图,对total列进行汇总计算,index为行,columns为列,margins=True增加一个全部行汇总,aggfunc=np.sum透视图中对total值进行sum计算,这里np是开头import的numpy as np,fill_value=0对空值进行0替换,否则没有数据会显示NaN
pv.head()
```

```python
pv.to_csv("f:/Pythonfortest/qiyeus9monthlytransactiontest2pv.csv")      #写入csv文件test2
```
